# to_csv

```python
df = pd.read_csv('ipl_deliveries.csv')
df.head()
```

```python
temp = df.groupby('batter')['batsman_run'].sum().reset_index()
```

```python
temp.to_csv('batsman_run.csv',index=False)
```

# to_excel

```python
temp.to_excel('bat.xlsx',index=False)
```

```python
temp.to_excel('bat.xlsx',sheet_name='First')
```

```python
temp.to_excel('bat.xlsx',sheet_name='Second')
```

```python
temp_1=df.pivot_table(index='batter',columns='BowlingTeam',values='batsman_run',aggfunc='sum')
```

**when you have ``multiple sheets``**

```python
with pd.ExcelWriter('output.xlsx') as writer:
    temp.to_excel(writer,sheet_name='First')
    temp_1.to_excel(writer,sheet_name='Second')
```

**``or``**

```python
writer = pd.ExcelWriter('final.xlsx')
temp.to_excel(writer,'original')
temp_1.to_excel(writer,'result')
writer.save()
```

# to_html

```python
df.query('batsman_run==6').pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count').to_html('sixes_heatmap.html')
```

# to_json

```python
df.groupby(['BattingTeam','batter'])['batsman_run'].sum().unstack().to_json('ipl.json')
```

```python
engine = create_engine('mysql+pymysql://root:@localhost/ipl')
# {root}:{password}@{url}/{database}
df.to_sql('ipl_delivery',con=engine,if_exists='append')
```
